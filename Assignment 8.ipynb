```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'X':[7, 2, 0, 3, 4, 2, 5, 0, 3, 4]})
```

```python
df
```

```python
pos_zero = list(i for i in range(len(df['X'])) if (df.iloc[i]['X']== 0))
```

```python
pos_zero
```

```python
count =[]
```

```python
pos = 0
```

```python
for i in range(len(df['X'])):
    if i in pos_zero:
        pos = 0
    else:
        pos = pos+1
    count.append(pos)
```

```python
df['Y']= count
```
